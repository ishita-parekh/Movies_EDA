```python
#Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib
plt.style.use('ggplot')

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# Read the data

df = pd.read_csv('C:/Users/ISHITA/Desktop/movies.csv')
```

```python
df
```

```python
# To check if there is any missing data

df.isnull().sum()
```

```python
# dropping missing data across rows, axis = 0

df.dropna(inplace=True)
```

```python
df
```

```python
df.dtypes
```

```python
# change data types of columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
# Sort the data by gross revenue

df = df.sort_values('gross',ascending=False)
```

```python
#To have no limit on how many rows to display - chnaging the default row values to display

#pd.set_option('display.max_rows',60)  
```

```python
# Drop duplicates

df.drop_duplicates()
```

```python
# To find correlation of budget with gross
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['gross'],y=df['budget'])
plt.title('Budget vs gross')
plt.xlabel('Gross')
plt.ylabel('Budget')
plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x=df['gross'],y=df['budget'],data=df,color="blue",scatter_kws={"color":"red"})
```

```python
# How much correlation between budget and gross

df.corr()
```

```python
# Plotting Correlation Matrix 

correlation_matrix = df.corr()
sns.heatmap(data=correlation_matrix,annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
#plt.show()
```

```python
# To find correlation of company
```

```python
# Converting Company column as numeric data to help in visulaization

df_numeric = df.copy()

for col in df_numeric.columns:
    if df_numeric[col].dtype == 'object':
        df_numeric[col] = df_numeric[col].astype('category')
        df_numeric[col] = df_numeric[col].cat.codes
        
df_numeric
```

```python
df
```

```python
#plotting correlation matrix after columns changed to categorical data

correlation_mat = df_numeric.corr()
sns.heatmap(data=correlation_mat,annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
# unstacking the correlation matrix

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
#sorting the matrix

sorted_corr = corr_pairs.sort_values()
sorted_corr
```

```python
#getting high correlation data

high_corr = sorted_corr[sorted_corr > 0.5]

high_corr
```
